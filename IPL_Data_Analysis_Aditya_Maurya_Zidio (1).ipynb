## Calculating strike rate of each batsman in IPL matches (2008-2024)

```python
# Assuming 'Striker' column contains the batsman's name
# and 'runs_scored' contains runs scored on each ball

# Grouping data by batsman to calculate total runs and balls faced
strike_rate_df = ballinfo.groupby('Striker').agg(
    Total_Runs=('runs_scored', 'sum'),  # Sum of runs scored by each batsman
    Balls_Faced=('Striker', 'count')    # Counting the number of balls faced
).reset_index()

# Calculating the strike rate: (Total Runs / Balls Faced) * 100
strike_rate_df['Strike_Rate'] = (strike_rate_df['Total_Runs'] / strike_rate_df['Balls_Faced']) * 100

# Sorting batsmen by highest strike rate in descending order
strike_rate_df = strike_rate_df.sort_values(by='Strike_Rate', ascending=False)

# Displaying the final DataFrame
print(strike_rate_df)
```

## Top 10 strikers in IPL matches (2008-2024)

```python
top_10_strikers = strike_rate_df.sort_values(by='Strike_Rate', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 5))
plt.barh(top_10_strikers['Striker'], top_10_strikers['Strike_Rate'], color='orange')

# Labels and title
plt.xlabel('Strike Rate')
plt.ylabel('Batsman')
plt.title('Top 10 Batsmen by Strike Rate (Minimum Balls Faced)')
plt.gca().invert_yaxis()  # Invert y-axis to have highest on top
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

## Hypotesting


### 1. Null hyptothesis : "There is no relationship between the venue and the match winner"


```python

from scipy.stats import chi2_contingency


# Create a contingency table for 'Venue' and 'Match Winner'
contingency_table = pd.crosstab(teamp['Venue'], teamp['Match_Winner'])

# Perform the Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi2 Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table: \n{expected}")

# Conclusion based on the p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a relationship between the venue and match winner.")
else:
    print("Fail to reject the null hypothesis: There is no relationship between the venue and match winner.")
```

### Output : Rejecting the null hypothesis. Hence , There is a relationship between the venue and match winner team.

### 2. Null hypothesis : There is no relationship between toss winner and the match winner

```python
# Create a contingency table for 'Venue' and 'Match Winner'
contingency_table = pd.crosstab(teamp['Toss_Winner'], teamp['Match_Winner'])

# Perform the Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi2 Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table: \n{expected}")

# Conclusion based on the p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a relationship between the toss winner and match winner.")
else:
    print("Fail to reject the null hypothesis: There is no relationship between the toss winner and match winner.")
```

### Output : Rejecting the null hypothesis. Hence, There is relationship between toss winner and the match winner
